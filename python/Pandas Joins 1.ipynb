```python
# right join
pd.merge(cust,ord, left_on = "ID",right_on = "C_ID" ,how = "right" )[["Name","Address","OID","Amount"]].query("Name == 'kaushik'").head(10)
```

```python
# inner join 
# keeping the selected columns before the head 
pd.merge(cust,ord, left_on = "ID",right_on = "C_ID" ,how = "inner" ).query("Name == 'kaushik'")[["Name","Address","OID","Amount"]].head(10)
```

```python
# to find is null values
pd.merge(cust,ord, left_on = "ID",right_on = "C_ID" ,how = "left" )[["Name","Address","OID","Amount"]].query("OID.isnull()").head(10)
```

```python
# to find not null 
pd.merge(cust,ord, left_on = "ID",right_on = "C_ID" ,how = "left" )[["Name","Address","OID","Amount"]].query("OID.notnull()").head(10)
```
