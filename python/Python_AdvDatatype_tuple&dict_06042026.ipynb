```python
# fromkeys
# setdefault

# questions : 
# can i insert a key/value pair in between ? {find alternate function to list insert}
#Not directly
original = {'a': 1, 'c': 3}
new_item = ('b', 2)
position = 1

# Convert items to a list to use the insert method
items = list(original.items())
items.insert(position, new_item)

# Re-create the dictionary
updated_dict = dict(items)
print(updated_dict)  # {'a': 1, 'b': 2, 'c': 3}

original = {'a': 1, 'c': 3}
new_item = ('b', 2)
position = 1

# Convert items to a list to use the insert method
items = list(original.items())
items.insert(position, new_item)

# Re-create the dictionary
updated_dict = dict(items)
print(updated_dict)  # {'a': 1, 'b': 2, 'c': 3}

# we are able to change value , but can we change keys ? {} Y/N 
#No but we can  delete and re add the key 
user_data = {'name': 'Alice', 'user_age': 30}

# We want to change 'user_age' to 'age'
user_data['age'] = user_data.pop('user_age')

print(user_data) # {'name': 'Alice', 'age': 30}
```
