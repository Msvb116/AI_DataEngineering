```python
import math

def add(a, b):
    print("The Addition of 2 numbers is", a + b)

def sub(a, b):
    print("The Subtraction of 2 numbers is", a - b)

def mul(a, b):
    print("The Multiplication of 2 numbers is", a * b)

def div(a, b):
    if b == 0:
        print("Error: Cannot divide by zero!")
    else:
        print("The Division of 2 numbers is", a / b)

def power(a, b):
    """Calculates a raised to the power of b (a^b)"""
    c = a ** b
    print(f"{a} raised to the power of {b} is {c}")

def modulus(a, b):
    """Returns the remainder of a division"""
    c = a % b
    print(f"The remainder (modulus) of {a} divided by {b} is {c}")

def floor_div(a, b):
    """Divides and rounds down to the nearest whole number"""
    c = a // b
    print(f"The floor division of {a} and {b} is {c}")

def sqrt(a):
    """Calculates the square root"""
    if a < 0:
        print("Error: Cannot take square root of a negative number!")
    else:
        c = math.sqrt(a)
        print(f"The square root of {a} is {c}")

def absolute(a):
    """Returns the distance from zero (removes negative sign)"""
    print(f"The absolute value of {a} is {abs(a)}")
```
