```python
# assignment
# write a sales function
# any sales < 5000 ---Bronze
# sales 5 k - 10k ----silver
# sales 10k - 15k ----gold
# > 15k           ----platinum

# in bike data one additional column need to add sales group 
```

```python
bike_data.shape
```

```python
def trans_grp(a):
    if a < 5000:
        return "Bronze"
    elif a > 5000 and a < 10000:
        return "Silver"
    elif a > 10000 and a < 15000:
        return "Gold"
    else:
        return "Platinum"
     
```

```python
bike_data['Sales_group'] = bike_data['Sales'].apply(trans_grp)
```
