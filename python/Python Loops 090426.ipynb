### Nested Loops 

```python
colours = ["Black" , "Blue", "Green"]
size = ["small" , "Medium", "Large"]
```

```python
for i in colours:
    for j in size:
        print(i,j)
```

```python
result= []
for i in colours:
    for j in size:
        result.append([i,j])
print(result)        
```
