```python
# can u write a reusable function # sales 
# sales < 5000 bronze 
# sales 5k - 10k  silver
# sales 10k - 15k gold
# > 15 k - platinum
```

```python
def sales_grp(s):
    if s > 0 and s < 5000:
        print("The Customer is Bronze")
    elif s > 5000 and s < 10000:
        print("The Customer is silver")  
    elif s >= 10000 and s < 15000:
        print("The Customer is Gold") 
    elif s >= 15000:
        print("The Customer is Platinum") 
    else:
        print("Invalid Sales") 
```

```python
sales_grp(9500)
```

```python
sales_grp(-89)
```
