```python
# I want to know all the years cars which is having  > 30 cars 
# Expected output (Year |  Total_Cars)
# is the ordering is important ?

data.groupby("Year")[["Car_Name"]].count() \
    .query("Car_Name > 30") \
    .sort_values("Car_Name", ascending = False) \
    .rename(columns={"Car_Name" : "Total_Cars"})  ------> Rename of column 
```

```python
data.groupby("Year")[["Car_Name"]].count() \
    .query("Car_Name > 30") \
    .rename(columns={"Car_Name" : "Total_Cars"}) \  ------> Rename of column 
    .sort_values("Total_Cars", ascending = False) 
    
```

```python
data.groupby("Year")[["Car_Name"]].count() \
    .rename(columns={"Car_Name" : "Total_Cars"}) \    ------> Rename of column 
    .query("Total_Cars > 30") \
        .sort_values("Total_Cars", ascending = False) 
```

```python
# i want to see all petrol cars for every year provided it has 25 cars

data.query("Fuel_Type == 'Petrol'").groupby("Year")[["Car_Name"]].count().query("Car_Name > 25 ")
#    This query is for filterning dataset                                 ---This query for filtering result         
    
```
