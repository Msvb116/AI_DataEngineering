#### How to structurise the learning?
#### 1) 1 column 1 aggregation 

```python
# one column one aggragation # order_data.groupby([ "Region"]).agg({"Sales":"sum"})

order_data.groupby("Category")[["Sales"]].sum() 
```

#### 2) 2 column 1 aggregation   # pass it in list of columns 

```python
order_data.groupby(["Category" , "Region"])[["Sales"]].sum()
```

#### 3) one column 2 aggregations 

```python
order_data.groupby("Region")[["Profit","Sales"]].sum()
```

#### 4) 2 aggregations 2 columns

```python
order_data.groupby(["Category" , "Region"])[["Sales","Profit"]].sum()
```

#### 5 ) Average sales of a column


```python
order_data.groupby("Region")[["Sales"]].mean()   # mean is average function
```

#### 5 a) One column 2 Aggregation  (one is sum one is avg)

```python
order_data.groupby([ "Region"]).agg({"Sales":"sum","Profit":"mean"}) 
```

#### 6) one column 1 aggregation with 2 diff calculations 
#### ex: for sales give sum and mean

```python
# order_data.groupby([ "Region"]).agg({"Sales":"sum","Sales":"mean"}) {This wont work bcz same keys for 2 diff values}

order_data.groupby([ "Region"]).agg({"Sales":["sum","mean"]})
```

#### 7) Rename column names 
#### syntax : .rename(columns= {"Sales" : "Total_Sales"})

```python
order_data.groupby("Category")[["Sales"]].sum().rename(columns= {"Sales" : "Total_Sales"})
```

```python
order_data.groupby([ "Region"])  \
        .agg({"Sales":["sum","mean"]}) \
        .rename(columns = {"sum" : "Total_Sales" ,"mean" : "Avg_Sales"}) 
        ##.sort_values("Total_Sales" , ascending = False)
```
